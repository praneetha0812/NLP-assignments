```python
import nltk
from nltk.tokenize import sent_tokenize
nltk.download('punkt')
```

```python
with open('data.txt', 'r', encoding='utf-8') as file:
    data = file.read()
```

```python
#a.total sentences
sentences = sent_tokenize(data)
sentences_count = len(sentences)
sentences_count
```

```python
#b.using split function on sentences to count words
words = []
for sentence in sentences:
    words.extend(sentence.split())

# Total number of words, it's also counts the emoticons as words, so that total words are as shown below.
word_count = len(words)
print("Number of tokens using split function are", word_count)
```

```python
#c. using word_tokenize, the total no of words have varied
from nltk.tokenize import word_tokenize
word = []
for sentence in sentences:
    word.extend(word_tokenize(sentence))

word_count = len(word)
print("Number of tokens using word_tokenize are", word_count)
```

```python
#d.lowercased all words and checking the tokens and types now
padam=[]
for sentence in sentences:
    padam.extend(word.lower() for word in word_tokenize(sentence))

# Total token's and the type's after lowercasing all the tokens
token_count = len(padam)
type_count = len(set(padam))

print("Number of tokens after lowecasing", token_count)
print("Number of types after lowercasing", type_count)
```

```python
#e  A dictionary to store word type counts
word_counts = {}
for sentence in sentences:
    for word in word_tokenize(sentence):
        word = word.lower()  # Convert to lowercase
        if word in word_counts:
            word_counts[word] += 1
        else:
            word_counts[word] = 1

# Sort the dictionary by frequency
sorted_word_counts = dict(sorted(word_counts.items(), key=lambda item: item[1], reverse=True))

# Print the sorted dictionary
for word, count in sorted_word_counts.items():
    print(f"{word}: {count}")
```

```python
#f. The most frequent word type
most_frequent_word = next(iter(sorted_word_counts))
most_frequent_count = sorted_word_counts[most_frequent_word]
print("Most frequent word type", most_frequent_word, "with count:", most_frequent_count)
```

```python
#g. The 5th most frequent word 'type'
fifth_most_frequent_word = list(sorted_word_counts.keys())[4]  # 0-indexed
fifth_most_frequent_count = sorted_word_counts[fifth_most_frequent_word]
print(" The 5th most frequent word type", fifth_most_frequent_word, "with count:", fifth_most_frequent_count)
```

```python
import matplotlib.pyplot as plt

# Extracting word frequencies and ranks
word_ranks = range(1, len(sorted_word_counts) + 1)
word_frequencies = list(sorted_word_counts.values())

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.plot(word_ranks, word_frequencies, color='blue', marker='o', linestyle='-')
plt.title("Zipf's Law: Word Frequencies vs. Ranks")
plt.xlabel("Word Ranks")
plt.ylabel("Word Frequencies")
plt.grid(True)
plt.yscale('log')  # Logarithmic scale for better visualization
plt.xscale('log')
plt.show()
```
